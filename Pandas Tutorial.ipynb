```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('SN_d_tot_V2.0.csv',delimiter=';',index_col=None,header=None,names=['Year','Month','Day','Date_dec','SN','u1','u2','u3'])
```

```python
df['Date']=pd.to_datetime(df[['Year','Month','Day']])
df.set_index('Date')
```

```python
df.plot(x='Date',y='SN')
```

```python
freq=np.fft.fftfreq(df.shape[0],d=1.) #dt= 1 day
fs=np.abs(np.fft.fft(df['SN']))/df.shape[0]
```

```python
period=(1./freq[freq>0])/365.
MaxFs=period[fs[freq>0].argmax()]
```

```python
plt.plot(period,fs[freq>0]) 
plt.vlines(MaxFs,0,fs[freq>0].max(),linestyle='--',linewidth=0.75,label='{:.2f} Years'.format(MaxFs))
plt.xlim(0,20)
plt.legend()
```
